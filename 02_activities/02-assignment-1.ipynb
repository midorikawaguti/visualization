```python
#Set up
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import PIL
import requests
import scipy

np.random.seed(613)
x=np.arange(50)
y1 = np.random.randint(0,100,50)
y2 = np.random.randint(0,100,50)
```

```python
#line plot with both of our y variables on the same axes
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y1)
ax.plot(x,y2)
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y1,
        label="Person 1")
ax.plot(x,y2,
        label="Person 2")
ax.legend(loc = 'lower right',
          frameon = True,
          fontsize=12,
          ncol = 2,
          shadow = True)
```

# Text and Annotations

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1,label = "Person 1")
ax.scatter(x, y2, label= "Person 2")
ax.legend(loc='lower right')

ax.text(10,95, "This value is important!")
```

### Modifying text annotations

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1,label = "Person 1")
ax.scatter(x, y2, label= "Person 2")
ax.legend(loc='lower right')

ax.text(10,95, "This value is important!",
        ha='center',
        color='red',
        size=20)
```

### Positioning text annotations

```python
fig, ax = plt.subplots()
ax.axis([0,10,0,10])

ax.text(1,5, "Data: (1,5)", transform = ax.transData)
ax.text(0.5, 0.1, "Axes: (0.5, 0.1)", transform = ax.transAxes)
ax.text(0.2, 0.2, "Figure: (0.2, 0.2)", transform = fig.transFigure)
```

### Annotation with arrows

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label="Person 2")
ax.legend(loc='lower right')

ax.annotate("This is important!", xy = (10,95), xytext = (20,94),
            arrowprops = dict(facecolor = 'black', shrink = 0.05))
```

## Axis Labels

```python
#Removing tick marks and labels

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1,label = "Person 1")
ax.scatter(x,y2, label="Person 2")
ax.legend(loc = "lower right")

ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())
```

```python

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1,label = "Person 1")
ax.scatter(x,y2, label="Person 2")
ax.legend(loc = "lower right")

ax.xaxis.set_major_locator(plt.MaxNLocator(3))
```

# Subplots

```python
#New Data
x1 = np.arange(50)
y1 = np.random.randint(0,75,50)
x2 = np.array(["Luffy", 'Zoro', 'Nami', 'Usopp', 'Sanji'])
y2 = np.array([110,180,240,99,220])
```

### Two plots next to each other

```python
fig, (ax1, ax2) = plt.subplots(ncols=2,
                nrows=1,
                figsize=(7,3))

ax1.scatter(x1, y1)
ax2.bar(x2, y2)
```

#### Customize plots

```python
fig, (ax1, ax2) = plt.subplots(ncols=2,
                nrows=1,
                figsize=(9,3))

ax1.scatter(x1, y1, color = 'green', marker = '*')
ax2.bar(x2, y2, color='pink' )
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                     ['ax2', 'ax3']],
                                     figsize = (7,4))

someaxes['ax1'].scatter(x1, y1)
someaxes['ax2'].bar(x2, y2)
someaxes['ax3'].plot(x1,y1)
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                     ['ax2', 'ax3']],
                                     figsize = (7,4))

someaxes['ax1'].scatter(x1, y1)
someaxes['ax2'].bar(x2, y2)
someaxes['ax3'].plot(x1,y1)

someaxes['ax1'].set_xlabel('A Big Label', fontsize = 18)
someaxes['ax2'].set_xlabel('Another Label', fontsize = 18)
someaxes['ax3'].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize=18)
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                     ['ax2', 'ax3']],
                                     figsize = (7,4),
                                     layout="constrained")

someaxes['ax1'].scatter(x1, y1)
someaxes['ax2'].bar(x2, y2)
someaxes['ax3'].plot(x1,y1)

someaxes['ax1'].set_xlabel('A Big Label', fontsize = 18)
someaxes['ax2'].set_xlabel('Another Label', fontsize = 18)
someaxes['ax3'].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize=18)
```

```python
y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")
```

### Add Error Information

```python
y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")

ax.errorbar(x, #our x values
y2, #our y values
yerr = y2_sd,
fmt = "none")
```

### Customize errorbar appearance

```python
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2,
        color = "red")

ax.errorbar(x,
            y2,
            yerr = y2_sd,
            fmt = "none",
            ecolor= "indigo",
            elinewidth= 4,
            capsize = 6,
            capthick= 4 )
```

### Errorevery

```python
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2,
        color = "red")
        
ax.errorbar(x,
            y2,
            yerr = y2_sd,
            fmt = "none",
            ecolor = "indigo",
            elinewidth = 4,
            capsize = 6,
            capthick = 4,
            errorevery= 2 )
```
