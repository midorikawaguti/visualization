## Basic scatterplot

```python
# create data
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y)
```

## Basic bar plot

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(x,y)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
```

## Histogram

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y)
```

## Adding axis labels and titles

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)

ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
```

## Modifying labels and titles

```python
font1={'family': 'sans-serif', 'color':'blue', 'size':20}
font2 = {'family':'monospace', 'color': 'green', 'size':14}

fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)

ax.set_title('Total growth over time', fontdict = font1)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
```

## Moving labels and titles

```python
font1={'family': 'sans-serif', 'color':'blue', 'size':20}
font2 = {'family':'monospace', 'color': 'green', 'size':14}

fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)

ax.set_title('Total growth over time', fontdict = font1, loc='left')
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
```

## Modifying data points

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y,
           marker = '*',
           color = 'indigo')
```

## Style and width of the line connecting points

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y,
           marker = '*',
           color = 'indigo',
           linestyle = '--',
           linewidth = 2)
```

## Modifying data points: colour

https://matplotlib.org/stable/gallery/color/named_colors.html#sphx-glr-gallery-color-named-colors-py

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y,
           marker = '*',
           color = '#7425b9',
           linestyle = '--',
           linewidth = 2,
           markersize = 12,
           markeredgecolor = '#fa9359',
           markerfacecolor = '#000000')
```

## Grid Lines

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y,
           marker = '*',
           color = '#7425b9',
           linestyle = '--',
           linewidth = 2,
           markersize = 12,
           markeredgecolor = '#fa9359',
           markerfacecolor = '#000000')
ax.grid(axis = 'x')
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y,
           marker = '*',
           color = '#7425b9',
           linestyle = '--',
           linewidth = 2,
           markersize = 12,
           markeredgecolor = '#fa9359',
           markerfacecolor = '#000000')
ax.grid()
```

## Modifying Grid Lines

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y,
           marker = '*',
           color = '#7425b9',
           linestyle = '--',
           linewidth = 2,
           markersize = 12,
           markeredgecolor = '#fa9359',
           markerfacecolor = '#000000')
ax.grid(axis = 'x',
        color = 'green',
        linestyle = '-.',
        linewidth = 2)
```

## Bubble plot


A bubble plot is very similar to a scatterplot. Using matplotlib library, a bubble plot can be constructed using the scatter() function. In the example, the following parameters are used:

x : The data position on the x axis
y : The data position on the y axis
s : The marker size
alpha : Transparancy ratio

```python
# create data
df = pd.DataFrame({
      'x': np.random.rand(40),
      'y': np.random.rand(40),
      'z': np.random.rand(40),
   })
df.head()

```

```python
plt.scatter(df['x'], df['y'], s=df['z']*1000, alpha=0.5)
plt.show()
```
